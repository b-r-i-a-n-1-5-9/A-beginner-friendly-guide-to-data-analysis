```python
# When calculating the spread of a sample, we utilize measures of dispersion.
# some of the measures of dispersion are standard deviation, interquartile range,and emperical variance.
# standard deviation measures how much a value has deviated from the mean
# standard deviation is the squareoot of the variance which can be obtained by [summation (object-mean)*2]
# Interquatile range devides the data into various quatiles, we have the first quatile which consist of 1/4 of the values and bellow
# the second quatile which is made up of half of the values .Q2 is also the median value.
# the third quatile consist of 3/4 of the values
# qwe also have deciles, which are devided into groups of 10, and centiles, which are grouped in terms of percentages because they are divisible by 100
# The interquatile range is the difference between the third quatile and the first quatile.
```

```python
# there are two ways of representing measures of dispersion. 
# the first one involves the use of histogram
# the other method involves the use of box plots. 
# box plots represents groups of numerical data through their quartiles . 
# box plots indicate the degree of dispersion in the data.
# the first quartile q1 and the third quartile q3 will determine the side of the box wherealse the median is in the middle.
# box plots then have whiskers, which are vertical lines that indicate the highest or the lowest values.
```

```python
#Now lets try to code

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data=pd.read_csv('operations.csv')
for i in data['categ'].unique():
    data_subsection=data[data.categ==i]
    print('_'*30)
    print (i)
    print('Mean:\n',data_subsection['amount'].mean())
    print('mode:\n',data_subsection['amount'].mode())
    print('median:\n',data_subsection['amount'].median())
    data_subsection['amount'].hist()
    plt.show()
    
```
