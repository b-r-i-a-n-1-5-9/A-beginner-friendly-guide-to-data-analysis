```python
# In data analysis, it is not efficient to only know the median, mean and mode
# at time, you have to know the shape of the data
# what do we mean by the shape? 
# we have to know if majority of the values are more than the mean, or less than the mean

#                             SKEWENESS 
    
    
# in the  realm of measuring the shape of our data, we come across a term known as skewness
# Skewness is the measure of assymetry or symetry of our data.
# In simple terms, this is an analysis of how our data is distributed around the central value.
# remember our measures of central tendancy, thats where we obtain the central value

# is skewness is equal to 0, it means that the distribution is perfectly symmetrical
# if the skewness is greater than 0, it means that our distribution is positively skewed, or as we say skewed to the right.
# if our distribution is less than 0, it means that the distribution is negatively skewed or skewed to the left. 
```

```python
# EMPIRICAL KURTOSIS 
# empirical kurtosis is the measure of peakedness or flatness of a distribution
# kurtosis analyses how a distribution differs from a normal distribution.
# It can be used to measure the extremity of  ourliers 
# in the real world, kurtosis can be used for risk management, financial modeling-----
# ----and other analysis where you need to measure the probability of the occurance of extreme values


# If the kurtosis is 0, it means that the distribution has the same degree of peakedness like a normal distribution. 
# this means that the distribution does not have extreme outliers

# if the kurtosis is greater than 0, it means that the distribution is more peaked, than the normal distribution.
# this shows that the distribution is densly concentrated around a certain value. and shows posible presence of outliers

# if the kurtosis is less than 0, it shows that the distribution is less peaked, or flatter than the normal distribution, 
# this means that the observations are less densely concentratedm and there is a lower likelyhood of outliers.
```

```python
# Before we move to the code part, it is good to know that when we say that we want to know whether the majority of the values are above or below the mean, we just mean that we want to know if the median is above  or bellow the mean 
# remember that the median is oftenly found in the middle and hence, if we want to determine which is greater, median or mean, we analyse the symetry of a distribution
# in a symetrical distribution, or where the shape on the left is similar to the shape on the right, it means that the mode is equal to the median and is equal to the mean
# symetrical distribution= Mode=Median=Mean
# if the distribution is skewed to the right, or it is positively skewed, it means that that the mean is greater than the median, which is then greater than the mode
# positive skeweness = Mode<Median<Mean
# if the distribution is skewed to the left, or negatively skewed, it means that Mode>Median>Mean
```

```python
import pandas  as pd
import numpy as np
import matplotlib.pyplot as plt

data=pd.read_csv('operations.csv')

for i in data['categ'].unique():
    subsection=data[data.categ==i]
    print('_'*30)
    print(i)
    print('Mean:\n',subsection['amount'].mean())
    print('Median:\n',subsection['amount'].median())
    print('Mode:\n',subsection['amount'].mode())
    print('Variance:\n',subsection['amount'].var(ddof=0))
    print('Standard deviation:\n',subsection['amount'].std(ddof=0))
    print('Skewedness:\n',subsection['amount'].skew())
    print('Kurtosis:\n',subsection['amount'].kurtosis())
    subsection['amount'].hist()
    plt.show()
    subsection.boxplot(column='amount',vert=False)
    plt.show()
    
```
