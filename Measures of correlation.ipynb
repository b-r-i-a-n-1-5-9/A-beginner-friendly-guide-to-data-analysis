```python
 #                       HOW DO WE USE A NUMERICAL INDICATOR TO ANALYSE CORELATION??

# We can alsoe use a numerical indicator to analyse the correnation between variables.
# when using numerical indicators, we want to know if when we have a low ballance X, we also have small transactions of the variable Y
# but then, what do we mean when we say small?
# we have to have a certain number which bellow it we can consider a numeric to be small and above it to be "not small"
# in this situation, our value of comparison will be the mean. 

# so, if we assume x to refer to the ballance before the transaction was made, and Y to reffere to the amount of transaction, 
# to measure whether x was small or great, we have to find the difference between x and the mean.

# so   ballance before transaction (x) - Mean value(x̄) [by mean value we mean the mean of the ballance before transaction  column]

# if ballance before transaction is bellow the mean, the answer to the above subtraction will be negative 

# if the ballance before transaction is above the mean, the answer to our subtraction will be positive 

# similar to the comparison between the amount of transaction
# Amount of transaction(y)- Mean value (Ȳ) [by mean value we mean the mean of the amounf of transaction column]

# if we subtract the mean valuey  from the amount of transaction  ie [Amount of transaction - Mean VALUE] 

# if the amount of transaction is less than the mean, the subtraction will lead to a negative value, 

# if the amount of transaction is greater than the mean, then the answer to our subtraction will be positive


# if we multiply the two answers together(the difference between ballance before transaction and mean value) and (the difference between amount of transaction and the maen  value) 

# if both of the answers between the two subtraction is a negative(are below the mean), the multiplication will result to a positive because two negative give a positive -(x-x̄)-(y-Ȳ)= positive value 

# if both of the answers of the subtractions are positive(are above the mean), the multiplication will be positive too (x-x̄)(y-Ȳ)= positive value 


# so, Assuming that the multiplication gives a value A
#                                         THERE EXISTS A CORRELATION

# if the all amounts are trully small when the ballance is low the subtraction will be negative, and the multiplication will be positive  -(x-x̄)-(y-Ȳ)
# the same applies to if the transaction are huge when the ballance before transaction is huge hence the multiplication will be positive   (x-x̄)(y-Ȳ) 

# the summation of the  mean of the above two sentences will be positive. -(x-x̄)-(y-Ȳ) /n +   (x-x̄)(y-Ȳ)/n = positive value 


#                                            THERE IS NO CORRELATION

# assuming that the values are spread all over, and there is no difinitive pattern , when one X value is low, the other Y value will not necesarily be low
# When X value is low, Y value will at times be low, and other times be high. so, the difference from the mean will be at times positive, and other times negative
# the multiplication between the differences from the mean value will also be negative at times, and positive at times
# so, the mean of the multiplication between the difference from the mean will be close to 0 because some values will be positive and other values negative, canceling each other compared to when the values are only positive


# SOOOOOOOO

# if x is low when Y is small and if x is great when Y is great, then the summation of the difference between the individual values and the mean will be positive . This will show a correlation
# If on the other hand they are not correlated, the difference between the individual values and the mean will be negative or positive and their multiplications will also be positive or negative. 
# the summation of the multiplication will be close to 0 



#                                             A NEGATIVE CORRELATION
# There is also the aspect og negative corelation

# this might occour when each and every time that x is high when Y is small and X is small and Y is great. after prerforming the above calculations, all the summations will provide a negative corelation
# this means that the two values are correlated, but an increase in one, leads to a decrease in the other. 

# one good analogy involves if x is academix grades, and Y is number of hours a person sleeps in class. X will be high if Y is small and X will be small if Y is great.


#                                            EMPIRICAL COVARIANCE  AND correlation coefficient 


# The ideology used in the above notebook give rise to an indicator we refer to as the Empiirical Covariance of X and Y

# if we want to normalize the Empirical covariance, we divide it by the product of the standard deviation of individual X values and Y values.

# The values obtained will be the Correlation Coefficient or the linear correlation coefficient or the Pearson's correlation coefficient



# We refer to the value as a lineir correlation because  the strategy only detects the relationship when they are linear

# when you normalize the Emperical covariance, corelated values will create a some what straight line. this shows that the coefficient is close to 1 or exual to 1  or close to -1 or e1ual to -1


# when items are not correlated, the emperical covariance will be close to 0

```

```python
#  Lets explain the code step by step 
import pandas as pd # inports the pandas library 
import numpy as np    # imports the numpy library
import matplotlib.pyplot as plt  # imports the matplotlib libarary


data=pd.read_csv('operations.csv')  # this reads the  CSV file named "operations" from our folder
display (data)                       # This displays the variable data which contains information from our operations data 

expenses=data[data.amount < 0]       # this creates a variable expenses which contains negative values that are below zero. 
# the above equation is esential because in our data, the amount column is made up of negative values 

plt.plot(expenses["balance_bef_trn"],-expenses['amount'],'o',alpha=0.5) # this plots a scatter plot 
# by default the function plot creates a line graph. However, it can also create a scatter plot if you specify using strings
# in the scatter plot the expenses["balance_bef_trn"] column will form the x axis and the -expenses['amount'] column will form the Y axis
#by using 'o' we are specifying the marker style  tell the code to return the plot function as a scatter points
# alpha=0.5 shows what the transparency level of the markets will be . an alpha value of 0.5 tells the code to create a semi-transparent marker.

plt.xlabel('Balance before Transactions')  # creates a title for the x axis labled Balance before transactions
plt.ylabel('expense amount')                # Creates a title for the Y axis labled expense amount. 
plt.show()                               # shows the scatter plot 
```

```python
# THE CORELATION COEFFICIENT & COVARIENCE MATRIX 

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

coefficient_corelation= st.pearsonr(expenses['balance_bef_trn'],-expenses['amount']) 
display(coefficient_corelation) 
covarience=np.cov(expenses['balance_bef_trn'],-expenses['amount'],ddof=0) 
display(covarience)
```
